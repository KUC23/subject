```python
검거율을 계산하고,
각 검거율 데이터를 colums에 추가

검거율은 검거율/발생*100
```

```python
crime_list = ['소계','살인','강도','강간','절도','폭력']
```

```python
# 데이터프레임에 존재하는 모든 변수를 리스트와 시킨 후 반복문을 이용하여 개별 검거율 계산과 동시에 열 가가

all_0 = 0

for i in crime_list:
    발생 = f'{i}(발생)'
    검거 = f'{i}(검거)'
    검거율 = f'{i}검거율'         

    print(발생,검거,검거율)
    df_excel_pivot[검거율] = (df_excel_pivot[검거]/df_excel_pivot[발생])*100      # 개별 검거율 추가
    all_0 += df_excel_pivot[검거율]
    print('전체 검거율',all_0)
    
df_excel_pivot['전체 검거율'] = all_0/len(crime_list)                             # 전체 검거율 추가
```

```python
# 개별 검거율 추가 후 전체 검거율 추가

df_excel_pivot
```

```python
rename_list_0 = ['강간(발생)','강도(발생)','살인(발생)','절도(발생)','폭력(발생)']
rename_list_1 = ['강간','강도','살인','절도','폭력']
```

```python
#원본 유지를 위한 복사본 작성

df_excel_pivot_rename = df_excel_pivot.copy()
```

```python
# 1. inplace=True 옵션으로 원하는 데이터 프레임의 원본을 변경하나나

for i in range(len(rename_list_0)):
    df_excel_pivot_rename.rename(columns={rename_list_0[i]: rename_list_1[i]},inplace=True)
```

```python
df_excel_pivot_rename
```

```python
#원본 유지를 위한 복사본 작성

df_excel_pivot_rename_1 = df_excel_pivot.copy()
```

```python
# 2. 새로운 변수를 만들어 원본을 유지하고, 복사본에 변수를 쌓아나감

for i in range(len(rename_list_0)):
    df_excel_pivot_rename_1 = df_excel_pivot_rename_1.rename(columns={rename_list_0[i]: rename_list_1[i]})
```
